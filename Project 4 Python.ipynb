# TIME SERIES ANALYSIS USING BOLLIGER BANDS AND EWMA Model

```python
# Close 20 MA
niftyit['Close: 20 Day Mean'] = niftyit['Close'].rolling(20).mean()

# Upper = 20MA + 2*std(20)
niftyit['Upper'] = niftyit['Close: 20 Day Mean'] + 2*(niftyit['Close'].rolling(20).std())

# Lower = 20MA - 2*std(20)
niftyit['Lower'] = niftyit['Close: 20 Day Mean'] - 2*(niftyit['Close'].rolling(20).std())

# Plot Close
niftyit[['Close','Close: 20 Day Mean','Upper','Lower']].plot(figsize=(20,10))

plt.show()
```

```python
# create 1 month SMA off of Close value
niftyit['30 Day SMA'] = niftyit['Close'].rolling(window=30).mean()

# plot SMA & Close Value
niftyit[['Close', '30 Day SMA']].plot(figsize=(10,8))
plt.show()
```

```python
# create EWMA
niftyit['EWMA-30'] = niftyit['Close'].ewm(span=30).mean()

# plot EWMA
niftyit[['Close', 'EWMA-30']].plot(figsize=(10,8))
plt.show()
```
